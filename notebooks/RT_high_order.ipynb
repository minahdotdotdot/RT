```julia
include("../codes/RT.jl")
seed=1205
Random.seed!(seed);
```

```julia
ϵ = 0.01;             # Nonlinear scale
ω = [-1, 3, -2];      # "slow" wave numbers
C = floatRT(5);       # Energy conserving constants
IC = onUnitCircle(3)  # Initial condition
T=1000;               # Final time
L=1000;               # Number of state/amplitudes to save
```

## True solution visualize

```julia
tsol = readdlm("../txtfiles/true_seed"*string(seed)*".txt")
using PyPlot
plot(tsol[:,1], label="z1")
plot(tsol[:,2], label="z2")
plot(tsol[:,3], label="z3")
legend()
```

```julia
using ExponentialUtilities
```

```julia
#IF_steppers = [IFE, IFRK2, IFRK3, IFRK4];
#ETD_steppers = [ETD1, ETDRK2, ETDRK3, ETDRK4, ETDRK4B]
#steppers=[IF_steppers;ETD_steppers];
steppers=[IFRK3, IFRK4, ETDRK3, ETDRK4, ARK3RT, ARK4RT]
ss = length(steppers)
h = 1. # Time-step
for j = 1 : ss
	txtname=string(steppers[j])*"h="*string(h)*"seed"*string(seed)
	N = Int(ceil(T/h)); 
    every = Int(ceil(N/L));
	RT_amp(N, h, every, IC, ω=ω, ϵ=ϵ, C=C, stepper=steppers[j], name=txtname);
end
```

```julia
solA = Array{Array{Float64,2},1}(undef,ss);
for i = 1 : ss
    solA[i] = readdlm("../txtfiles/"*string(steppers[i])*"h="*string(h)*"seed"*string(seed)*".txt")
end
```

```julia
for i = 1 : 6
    subplot(320+i)
    semilogy(abs.(tsol[:,1]-solA[i][:,1]), label="z1")
    semilogy(abs.(tsol[:,1]-solA[i][:,2]), label="z2")
    semilogy(abs.(tsol[:,1]-solA[i][:,3]), label="z3")
    title(string(steppers[i])*", h ="*string(h))
    #legend()
end
tight_layout(pad=1.0)
```

```julia
h = 2. # Time-step
for j = 1 : ss
	txtname=string(steppers[j])*"h="*string(h)*"seed"*string(seed)
	N = Int(ceil(T/h)); 
    every = Int(ceil(N/L));
	RT_amp(N, h, every, IC, ω=ω, ϵ=ϵ, C=C, stepper=steppers[j], name=txtname);
end
```
