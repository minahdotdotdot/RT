## Padé Approximation of the exponential function

The Taylor series expansion of $e^x$ has the same coefficients for real and complex $x$'s.

We define $P_{M}^N(x) = \frac{a_0+a_1 x + a_2 x^2 + \cdots + a_N x^N}{1 + b_1 x + b_2 x^2 + \cdots + b_M x^M}$. 

There are details notes in ``../notes/Pade.pdf``, and we denote $P_{M}^N(x)$ in the plots with ``Pade N-M``.

```julia
include("../codes/pade.jl")
```

#### Cases where m=n
Here, we look at Padé approximations for when the rational functions have the same degree polynomails in the numerator and denominator.

```julia
x = range(0, pi/2, length=1001)
for i = 2: 8
    m = i; n=i;
    c = expω(2*i) #This returns the TS coefficients for exp(x).
    plotPade!(c, n, m, x)
end
xt = [0, .125*pi, .25*pi, .375*pi, .5*pi]#, 3*pi, 4*pi]
axvline.(xt[2:end], color="k")
xlabels = ("0", L"\frac{\pi}{8}",  L"\frac{\pi}{4}", L"\frac{3\pi}{8}", L"\frac{\pi}{2}")#, L"3\pi", L"4\pi")
xticks(xt, xlabels)
legend()
title("Pade approx errors of exp(ix)")
```

Let's also look at how these approximations behave beyond a full period. 

```julia
x = range(0, 4*pi, length=1001)
for i = 2: 8
    m = i; n=i;
    c = expω(2*i) #This returns the TS coefficients for exp(x).
    plotPade!(c, n, m, x)
end
xt = [0, pi, 2*pi, 3*pi, 4*pi]#, 3*pi, 4*pi]
axvline.(xt[2:end], color="k")
xlabels = ("0", L"\pi",  L"2\pi", L"3\pi", L"4\pi")#, L"3\pi", L"4\pi")
xticks(xt, xlabels)
legend()
title("Pade approx errors of exp(ix)")
```

#### Cases where m+n = 8
Let's now see if the degrees of the denominator and numerator polynomials matter, given the same degrees of freedom in calculating the coefficients. 

```julia
x = range(0, pi/2, length=1001)
for i = 2: 6
    m = i; n=8-i
    c = expω(8) #This returns the TS coefficients for exp(x).
    plotPade!(c, n, m, x)
end
xt = [0, .125*pi, .25*pi, .375*pi, .5*pi]#, 3*pi, 4*pi]
axvline.(xt[2:end], color="k")
xlabels = ("0", L"\frac{\pi}{8}",  L"\frac{\pi}{4}", L"\frac{3\pi}{8}", L"\frac{\pi}{2}")#, L"3\pi", L"4\pi")
xticks(xt, xlabels)
legend()
title("Pade approx errors of exp(ix)")
```
